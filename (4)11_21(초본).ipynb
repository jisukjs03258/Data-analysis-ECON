# Returns

#### One-period return


  - one period simple return -> pct_change()

  $$ R_t=\frac{p_{t}-p_{t-1}}{p_{t-1}}$$

- one-period gross return ->

  $$ \frac{p_{t}}{p_{t-1}}=R_t+1$$

#### Multiple period return
- holding period가 1단위 이상일 때

**k-period simple return from time $t-k$ to $k$**

 $$R_t(k)=\frac{p_t-p_{t-k}}{p_{t-k}}$$

- k-period gross return
  $$\frac{p_{t}}{p_{t-k}}=R_t(k)+1$$


-----

  $$1+R_t(k)=(1+R_t)\times(1+R_{t-1})\times\cdots \times (1+R_{t-k1})$$

  $$R_t(k)=(1+R_t)\times(1+R_{t-1})\times\cdots \times (1+R_{t-k1})-1$$

```python
((data.pct_change()+1).cumprod()).plot() #cumprod 각각의 데이터들을 다 곱해라
# title : Gross return
plt.show()
```
