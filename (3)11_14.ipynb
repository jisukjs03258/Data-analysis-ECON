### Data normalization

- Min-Max

$$x_{mm}=\frac{x-x_{min}}{x_{max}-x_{min}}$$

특정값의 min과 max의 차이가 너무 클때 사용(??)

```python
df["length"].head()
```

```python
df["length"]=(df["length"]-df["length"].min())/(df["length"].max()-df["length"].min())
```

```python
df["length"] #length가 줄어들었음
```

```python
df["length"]=(df["length"]-df["length"].mean())/df["length"].std()
#z score
```

## Binning

- 연속적 수치형 데이터를 비연속적인 범주형 데이터로 변환

```python
df["horsepower"].unique() #104.22277... 이거만 float. int가 아님
```

```python
len(df["horsepower"].value_counts())
```

```python
df["horsepower"]=df["horsepower"].astype(int, copy=True) #copy가 True여야 실수도 int로 바뀜!!!!!
```

```python
df["horsepower"].unique() #104.222가 104로 바뀜
```

```python
df["horsepower"].value_counts()
```

```python
plt.hist(df["horsepower"])
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower")
plt.show()
```

```python
list_horsepower=df["horsepower"].unique().tolist()
```

```python
min(list_horsepower)
```

```python
max(list_horsepower)
```

```python
bins=np.linspace(min(df["horsepower"]),max(df["horsepower"]), 4) #min값과 max값 사이를 4개로 linspace!
```

```python
bins
```

- pd.cut(X, bins, labels)메소드를 활용하여 연속형 변수의 여러개 구간별 범주화 시도

```python
group_name=["low","medium","high"]
```

```python
df["horsepower-binned"]=pd.cut(df["horsepower"], bins, labels=group_name, include_lowest=True)
```

```python
df["horsepower-binned"]
```

```python
df["horsepower-binned"].value_counts() #구간별 숫자를 통일시키는 게 아니라,기준에 맞으면 다 짤라!|
```

```python
plt.bar(group_name, df["horsepower-binned"].value_counts(),color="pink")
plt.xlabel=("horsepower")
plt.ylabel=("count")
plt.title("horsepower bins")
plt.show()
```
